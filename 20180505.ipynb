```python
random_index = np.random.permutation(178)
```

```python
random_index
```

```python
train_index = random_index[:140]
test_index = random_index[140:]
```

```python
178 * 0.8
```

```python
from keras.utils import np_utils
```

```python
train_dataset = df.iloc[train_index]
test_dataset = df.iloc[test_index]
```

```python
train_label = np_utils.to_categorical(train_dataset['label'] - 1)
test_label = np_utils.to_categorical(test_dataset['label'] - 1)
```

```python
train_label.shape
```

```python
cols
```

```python
train_feature = (train_dataset[cols].values - tmp_mean) / tmp_std
test_feature = (test_dataset[cols].values - tmp_mean) / tmp_std
```

```python
from keras.layers import Input
from keras.layers.core import Dense
from keras.optimizers import SGD
from keras.models import Model
```

```python
inp = Input(shape = (13, ))
oup = Dense(3, activation = 'softmax')(inp)
model = Model (inputs = inp, outputs = oup)
```

```python
sgd = SGD(lr = 0.1)
model.compile(loss = 'categorical_crossentropy', optimizer = sgd)
```

```python
model.fit(train_feature, train_label, batch_size = 5, 
          epochs = 1, validation_data = (test_feature, test_label))
```

```python
np.argmax(model.predict(test_feature), axis = 1) == np.argmax(test_label, axis = 1)
```

```python
np.mean(np.argmax(model.predict(test_feature), axis = 1) == np.argmax(test_label, axis = 1))
```

```python
for i in [0, 1, 2]:
    preds = model.predict(test_feature)[:, i]
    labels = test_label[:, i]
    print(i, roc_auc_score(labels, preds), average_precision_score(labels, preds))
```

```python
import xgboost as xgb
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('wine.data', sep = ',', header = None)

indexes = np.random.permutation(df.shape[0])
pos = int(len(indexes) * 0.8)

train_indexes = indexes[:pos]
test_indexes = indexes[pos:]

def to_label_feature(tdf):
    label = tdf[0].values - 1
    feature = tdf[tdf.columns[1:]].values
    return label, feature
```

```python
y_train, x_train = to_label_feature(df.iloc[train_indexes])
y_test, x_test = to_label_feature(df.iloc[test_indexes])
```

```python
num_trees = 50
g_train = xgb.DMatrix(x_train, label = y_train)
g_test = xgb.DMatrix(x_test, label = y_test)
bst = xgb.train({'objective' : 'multi:softprob',
                'eta' : 0.1,
                'max_depth' : 6,
                'num_class' : 3,
                'lambda' : 0.1,
                'alpha' : 0.1}, g_train, num_trees)
```

```python
pred = bst.predict(g_test)

print np.argmax(pred, 1)

print y_test

print 'accuracy', np.mean(np.argmax(pred, 1) == y_test)
```
